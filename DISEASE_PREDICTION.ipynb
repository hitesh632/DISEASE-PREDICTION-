# PRE PROCESSING

```python
df.isna().sum()
df.isnull().sum()

cols = df.columns
data = df[cols].values.flatten()

s = pd.Series(data)
s = s.str.strip()
s = s.values.reshape(df.shape)

df = pd.DataFrame(s, columns=df.columns)

df = df.fillna(0)
df.head()
```

```python
print(df1.isnull().sum())
```

```python
vals = df.values
vals
```

```python
symptoms = df1['Symptom'].unique()
symptoms
```

```python
for i in range(len(symptoms)):
    vals[vals == symptoms[i]] = df1[df1['Symptom'] == symptoms[i]]['weight'].values[0]
    
d = pd.DataFrame(vals, columns=cols)
d.head()
```

```python
d = d.replace('dischromic _patches', 0)
d = d.replace('spotting_ urination',0)
df = d.replace('foul_smell_of urine',0)
df.head()
```

```python
(df[cols] == 0).all()  ## GETTING COLOUMNS WITH 0 AS A ENTITY


df['Disease'].value_counts() ## GETTING VALUE COUNTS OF ALL DISEASE

df['Disease'].unique()  ## GETTING ALL THE DISEASE

data = df.iloc[:,1:].values  # storing all the weights of symptoms as data


labels = df['Disease'].values #storing disease as labels
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data, labels, shuffle=True, train_size = 0.85)
#print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)


```

# MODEL BUILDING

```python
from sklearn.svm import SVC

model = SVC()
model.fit(x_train, y_train)

preds = model.predict(x_test)
print(preds)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(y_test,preds)
print(cfm)

print('classification report: ',)
print(classification_report(y_test,preds))

acc=accuracy_score(y_test,preds)
print('accuracy of the model: ',acc)
```
